```python
#  IMPORTS AND OOP SETUP
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
import os

# Add src to path
sys.path.append('../src')

from analysis_engine import FinancialNewsAnalyzer, TechnicalAnalyzer

print(" OOP classes imported successfully!")
```

```python
#  INITIALIZE ANALYZERS
news_analyzer = FinancialNewsAnalyzer()
print(" FinancialNewsAnalyzer initialized")

# Load sample stock data for technical analysis
stock_df = pd.read_csv('../data/Data/AAPL.csv')
stock_df['Date'] = pd.to_datetime(stock_df['Date'])
tech_analyzer = TechnicalAnalyzer(stock_df)
print(" TechnicalAnalyzer initialized")
```

```python
#  LOAD DATA USING OOP
print(" Loading data with OOP methods...")
news_analyzer.load_news_data('../data/raw_analyst_ratings.csv')
news_analyzer.load_stock_data('AAPL', '../data/Data/')
print(" Data loaded successfully!")
```

```python
#  RUN CORRELATION ANALYSIS WITH OOP
print(" Running OOP correlation analysis...")

# Merge datasets
merged_data = news_analyzer.merge_datasets('AAPL')
print(f" Merged {len(merged_data)} records")

# Calculate correlations
correlations = news_analyzer.calculate_correlations()
print(f" Pearson correlation: {correlations['pearson']['correlation']:.4f}")

# Generate report
report = news_analyzer.generate_report()
print(" Report generated")

# Technical analysis
tech_analyzer.calculate_sma(20)

tech_analyzer.calculate_rsi(14)
signals = tech_analyzer.get_trading_signals()

print(f" Trading signals: {signals}")
print(" OOP Analysis Complete!")
```
